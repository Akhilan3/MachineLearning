```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.pandas.set_option('display.max_columns',None)
import warnings
warnings.filterwarnings('ignore')


```

```python
car1=pd.read_csv("/tc20171021.csv",on_bad_lines='skip')
car1.head(10)
```

```python
car1.drop(labels=['Unnamed: 9','Unnamed: 10'],axis = 1,inplace=True)
```

```python
car2=pd.read_csv("/true_car_listings.csv",on_bad_lines='skip')
car2.head(10)
```

```python
 car2.shape
```

```python
 cars = pd.concat([car1,car2],axis=0)
```

```python
cars.drop('Id',axis=1,inplace=True)
```

Numerical Features

```python
cars['Price'].value_counts()
```

```python
sns.scatterplot(x=cars.Year,y=cars.Price)
```

```python
cars.groupby('Year')['Price'].mean().plot()
plt.xlabel('Year')
plt.ylabel('Mean Sale Price')
plt.title(" Price vs Year")
```

```python
cars.groupby('Year').aggregate({'Price':'mean'}).plot.bar()
```

Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encode = LabelEncoder()
```

```python
cars['city_encode'] = label_encode.fit_transform(cars['City'])
cars['State_encode'] = label_encode.fit_transform(cars['State'])
cars['Make_encode'] = label_encode.fit_transform(cars['Make'])
cars['Model _encode'] = label_encode.fit_transform(cars['Model'])
```

```python
cars.head(10)
```

```python
cars.drop(['City','State','Make','Model'],axis=1,inplace=True)
```

```python
cars.head(10)
```

```python
plt.figure(figsize=(12,6))
corr_matrix = cars.corr()
sns.heatmap(corr_matrix,annot=True,cmap='BuPu')
plt.show()
```
